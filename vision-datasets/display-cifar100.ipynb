```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
```

```python
transform = transforms.Compose([transforms.ToTensor()])
trainset = torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=False)
images, labels = next(iter(trainloader))
```

```python
fig, axes = plt.subplots(2, 7, figsize=(10, 3))  # 2x7 grid
for i in range(14):
    row = i // 7  # det the row of the subplot
    col = i % 7   # det the column of the subplot
    img = images[i].numpy().transpose((1, 2, 0))
    # display the image in the corresponding subplot
    axes[row, col].imshow(img)
    axes[row, col].set_title(trainset.classes[labels[i]])
    axes[row, col].axis('off')  # turn off axis numbers and ticks
plt.show()
```
